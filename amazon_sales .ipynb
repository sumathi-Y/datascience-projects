```python
#Connect method-To connect MySQL database
import pymysql
connection=pymysql.connect(
    host='localhost',
    user='root',
    password='mysql1234',
    db='amazon',
)
#Cursor method-Start pointing to the database,make connection to execute the query
cursor=connection.cursor()
sales_gender="SELECT Branch , Gender, count(*) as gender_count FROM amazon_sales GROUP BY Branch,Gender order by branch;"
#Execute method-To Execute the query through Python
cursor.execute(sales_gender)
#Fetchall method-To fetch the output of whole SQL query executed
output=cursor.fetchall()
#To get the output
print(output)
```

```python
#creating dataframe for analyses
df_gender=pd.read_sql(sales_gender,connection)
df_gender.head(6)
```

```python
plt.figure(figsize=(10,4))
ax=sns.barplot(x = 'Branch',
               y = 'gender_count',
               hue = 'Gender',
               data = df_gender)
ax.bar_label(ax.containers[0])
# Show the plot
plt.show()
```

```python
#Connect method-To connect MySQL database
import pymysql
connection=pymysql.connect(
    host='localhost',
    user='root',
    password='mysql1234',
    db='amazon',
)
#Cursor method-Start pointing to the database,make connection to execute the query
cursor=connection.cursor()
revenue="SELECT Product_line , city, customer_type, ROUND(SUM(total),2) as total_revenue FROM amazon_sales GROUP BY Product_line , city, customer_type;"
#Execute method-To Execute the query through Python
cursor.execute(revenue)
#Fetchall method-To fetch the output of whole SQL query executed
output=cursor.fetchall()
#To get the output
print(output)
```

```python
#creating dataframe for analyses
df_revenue=pd.read_sql(revenue,connection)
df_revenue.head(6)
```

```python
plt.figure(figsize=(15,7))
ax=sns.barplot(x = 'Product_line',
               y = 'total_revenue',
               hue = 'customer_type',
               data = df_revenue)
ax.bar_label(ax.containers[0])
# Show the plot
plt.show()
```

```python
plt.figure(figsize=(15,7))
ax=sns.barplot(x = 'city',
               y = 'total_revenue',
               hue = 'customer_type',
               data = df_revenue)
ax.bar_label(ax.containers[0])
# Show the plot
plt.show()
```
